```python
import pandas as pd
df=pd.read_csv("data/student_performance.csv")
#Data Cleansing
# Data Understanding & Inspection
print(f"Df Shape : {df.shape}")
print("Column Names : ")
print(df.columns)
print("Column Information")
df.info()
Selected_df=df.copy()
#Noise Reduction

Selected_df=Selected_df.drop(columns=["student_id","gender","school_type","parental_education","internet_access","weekend_alcohol_1_5","family_support","free_time_1_5","passed","extracurricular",	"health_1_5"	,"weekday_alcohol_1_5"]).copy()
```

```python
Selected_df.describe()
'''
We are going to perform analysis to check Student Performance Factor 

'''
```

```python
# Handling Missing Values
print(f" No of Rows contain Null values :\n { Selected_df.isnull().sum()}")
```

```python
Selected_df.dropna(subset=["final_grade_0_100"],inplace=True)

Selected_df["age"]=Selected_df["age"].fillna(Selected_df["age"].mean()).round(2)
Selected_df["study_time_hours_per_week"]=Selected_df["study_time_hours_per_week"].fillna(Selected_df["study_time_hours_per_week"].mean()).round(2)
Selected_df["failures"]=Selected_df["failures"].fillna(Selected_df["failures"].mean()).round(2)
Selected_df["absences"]=Selected_df["absences"].fillna(Selected_df["absences"].mean()).round(2)
Selected_df["math_score_0_100"]=Selected_df["math_score_0_100"].fillna(Selected_df["math_score_0_100"].mean()).round(2)
Selected_df["reading_score_0_100"]=Selected_df["reading_score_0_100"].fillna(Selected_df["reading_score_0_100"].mean()).round(2)
Selected_df["writing_score_0_100"]=Selected_df["writing_score_0_100"].fillna(Selected_df["writing_score_0_100"].mean()).round(2)

print(f" No of Rows contain Null values :\n { Selected_df.isnull().sum()}")
Selected_df.head()

```

```python
#Removing Duplicates
print(f"Check Duplicates if exists : {Selected_df.duplicated().sum()}")
```

```python
#Data Type Correction
print(" Check if Data types need conversion ")
print(Selected_df.dtypes)
```

```python
numeric_cols=Selected_df.select_dtypes(include="number").columns
for cols in numeric_cols:
    Q1=Selected_df[cols].quantile(0.25)
    Q3=Selected_df[cols].quantile(0.75)
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    outlier=Selected_df[(Selected_df[cols]<lower)|(Selected_df[cols]>upper)]
    print(f"So the Outliers we have in {cols}  is {outlier.shape[0]}")
```

```python
#if outliers are data errors
#Data Validation
for_age=Selected_df.loc[(Selected_df["age"]<3)|(Selected_df["age"]>40),["age"]]
display(for_age)
for_study_time_hours_per_week=Selected_df.loc[(Selected_df["study_time_hours_per_week"]<0)|(Selected_df["study_time_hours_per_week"]>80),["study_time_hours_per_week"]]
display(for_study_time_hours_per_week)
for_failures=Selected_df.loc[(Selected_df["failures"]<0)|(Selected_df["failures"]>10),["failures"]]
display(for_failures)
for_absences=Selected_df.loc[(Selected_df["absences"]<0)|(Selected_df["absences"]>40),["absences"]]
display(for_absences)
for_math_score_0_100=Selected_df.loc[(Selected_df["math_score_0_100"]<0)|(Selected_df["math_score_0_100"]>100),["math_score_0_100"]]
display(for_math_score_0_100)
for_reading_score_0_100=Selected_df.loc[(Selected_df["reading_score_0_100"]<0)|(Selected_df["reading_score_0_100"]>100),["reading_score_0_100"]]
display(for_absences)
for_writing_score_0_100=Selected_df.loc[(Selected_df["writing_score_0_100"]<0)|(Selected_df["writing_score_0_100"]>100),["writing_score_0_100"]]
display(for_writing_score_0_100)
for_final_grade_0_100=Selected_df.loc[(Selected_df["final_grade_0_100"]<0)|(Selected_df["final_grade_0_100"]>100),["final_grade_0_100"]]
display(for_final_grade_0_100)
print("None of the outlier is data error")
```
